# Plotting CO$_2$ data from Mauna Loa Observatory

Data taken from [NOAA ESRL](https://www.esrl.noaa.gov/gmd/ccgg/trends/)

```python
import matplotlib.pyplot as plt
from read_co2 import read_co2

```

```python
# Make the plots a bit bigger to see
# NOTE: Must be done in a separate cell
plt.rcParams['figure.dpi'] = 100
```

```python
# Get the CO2 data
dates,data = read_co2('co2_mm_mlo.txt')



# Plot it
plt.plot( dates, data )
plt.title(r"Atmospheric CO$_{2}$ at Mauna Loa Observatory")
plt.xlabel("Date")
plt.ylabel(r"CO$_{2}$ Concentration, ppm")

plt.show()
```
